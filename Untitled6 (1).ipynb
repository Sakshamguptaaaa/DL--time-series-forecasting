```python

# =========================
# 1. IMPORTS
# =========================
import numpy as np
import pandas as pd
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
from sklearn.metrics import mean_absolute_error, mean_squared_error
import math

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")


# =========================
# 2. PERSONALIZED PARAMETERS
# =========================

roll_number = "102317256"

# WHY: parameters derived from roll number as per assignment constraint
window_size = (1+0+2+3+1+7+2+5+6) % 10 + 8
prediction_horizon = (56 % 3) + 1
hidden_size = (102 % 16) + 8

print(window_size, prediction_horizon, hidden_size)


# =========================
# 3. WINDOWING FUNCTION
# =========================

def create_windows(data, window_size, horizon):
    """
    WHY:
    Converts time-series into supervised learning format.
    Model cannot understand raw sequence → needs (input → output) pairs.
    """
    X, y = [], []

    for i in range(len(data) - window_size - horizon + 1):
        X.append(data[i:i+window_size])
        y.append(data[i+window_size:i+window_size+horizon])

    return np.array(X), np.array(y)


# =========================
# 4. DATASET 1 (CUSTOM)
# =========================

# Simple sine wave (clean pattern → helps explanation)
t = np.arange(0, 200, 0.1)
data1 = np.sin(t)

# normalize
data1 = (data1 - data1.mean()) / data1.std()

X1, y1 = create_windows(data1, window_size, prediction_horizon)

# chronological split (IMPORTANT)
split = int(0.8 * len(X1))
X1_train, X1_test = X1[:split], X1[split:]
y1_train, y1_test = y1[:split], y1[split:]


# =========================
# 5. DATASET 2 (ELECTRICITY)
# =========================

# Upload CSV in Colab OR give path
# df = pd.read_csv('/content/electricity.csv')

# For safety fallback (if dataset not loaded)
df = pd.DataFrame({'value': data1})  # replace with real dataset

series = df.iloc[:, 0].values

series = (series - np.mean(series)) / np.std(series)

X2, y2 = create_windows(series, window_size, prediction_horizon)

split2 = int(0.8 * len(X2))
X2_train, X2_test = X2[:split2], X2[split2:]
y2_train, y2_test = y2[:split2], y2[split2:]


# =========================
# 6. TENSOR CONVERSION
# =========================

def to_tensor(X, y):
    X = torch.tensor(X, dtype=torch.float32).unsqueeze(-1)  # (batch, seq, 1)
    y = torch.tensor(y, dtype=torch.float32)
    return X.to(device), y.to(device)

X1_train, y1_train = to_tensor(X1_train, y1_train)
X1_test, y1_test = to_tensor(X1_test, y1_test)

X2_train, y2_train = to_tensor(X2_train, y2_train)
X2_test, y2_test = to_tensor(X2_test, y2_test)


# =========================
# 7. MLP BASELINE
# =========================

class MLP(nn.Module):
    def __init__(self):
        super().__init__()
        self.fc1 = nn.Linear(window_size, hidden_size)
        self.relu = nn.ReLU()
        self.fc2 = nn.Linear(hidden_size, prediction_horizon)

    def forward(self, x):
        # WHY: flatten removes temporal structure → baseline comparison
        x = x.view(x.size(0), -1)
        x = self.relu(self.fc1(x))
        return self.fc2(x)


# =========================
# 8. CUSTOM RNN (MANDATORY)
# =========================

class CustomRNN(nn.Module):
    def __init__(self):
        super().__init__()
        self.hidden_size = hidden_size

        self.Wx = nn.Linear(1, hidden_size)
        self.Wh = nn.Linear(hidden_size, hidden_size)
        self.fc = nn.Linear(hidden_size, prediction_horizon)

    def forward(self, x):
        # WHY: hidden state stores memory across time steps
        h = torch.zeros(x.size(0), self.hidden_size).to(device)

        for t in range(x.size(1)):
            # WHY: combine current input + previous memory
            h = torch.tanh(self.Wx(x[:, t]) + self.Wh(h))

        return self.fc(h)


# =========================
# 9. TRAIN FUNCTION
# =========================

def train_model(model, X, y, epochs=20):
    optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
    loss_fn = nn.MSELoss()

    losses = []

    for epoch in range(epochs):
        model.train()
        optimizer.zero_grad()

        preds = model(X)
        loss = loss_fn(preds, y)

        loss.backward()
        optimizer.step()

        losses.append(loss.item())

    return losses


# =========================
# 10. EVALUATION
# =========================

def evaluate(model, X, y):
    model.eval()
    with torch.no_grad():
        preds = model(X).cpu().numpy()
        y_true = y.cpu().numpy()

    mae = mean_absolute_error(y_true, preds)
    mse = mean_squared_error(y_true, preds)
    rmse = math.sqrt(mse)

    return preds, mae, mse, rmse


# =========================
# 11. TRAIN + PLOT FUNCTION
# =========================

def run_pipeline(X_train, y_train, X_test, y_test, title):

    model = CustomRNN().to(device)

    losses = train_model(model, X_train, y_train)

    preds, mae, mse, rmse = evaluate(model, X_test, y_test)

    print(f"{title} → MAE:{mae:.4f}, MSE:{mse:.4f}, RMSE:{rmse:.4f}")

    # LOSS PLOT
    plt.figure()
    plt.plot(losses)
    plt.title(f"{title} Training Loss")
    plt.show()

    # PREDICTION PLOT
    plt.figure()
    plt.plot(y_test.cpu().numpy()[:100, 0])
    plt.plot(preds[:100, 0])
    plt.title(f"{title} Prediction vs Actual")
    plt.legend(["Actual", "Predicted"])
    plt.show()

    return mae, mse, rmse


# =========================
# 12. RUN MAIN EXPERIMENT
# =========================

print("DATASET 1")
run_pipeline(X1_train, y1_train, X1_test, y1_test, "Dataset 1")

print("DATASET 2")
run_pipeline(X2_train, y2_train, X2_test, y2_test, "Dataset 2")


# =========================
# 13. ABLATION STUDY
# =========================

def ablation(data, name):

    for factor in [0.5, 1, 2]:
        w = int(window_size * factor)

        if w < 5:
            continue

        X, y = create_windows(data, w, prediction_horizon)

        split = int(0.8 * len(X))
        X_train, X_test = X[:split], X[split:]
        y_train, y_test = y[:split], y[split:]

        X_train, y_train = to_tensor(X_train, y_train)
        X_test, y_test = to_tensor(X_test, y_test)

        print(f"\n{name} | Window = {w}")
        run_pipeline(X_train, y_train, X_test, y_test, f"{name} (w={w})")


print("\nABLATION DATASET 1")
ablation(data1, "Dataset1")
```
